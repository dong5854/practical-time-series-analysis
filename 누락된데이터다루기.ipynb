```python
import pandas as pd

df : pd.DataFrame = pd.read_excel('./data/Labor Force Statistics from the Current Population Survey.xlsx', skiprows=range(0,11), index_col=0)
df
```

```python
# 임의로 누락된 데이터로 구성된 데이터셋을 생성합니다.
rand_df : pd.DataFrame = df.sample(n=5)
```

```python
rand_df: pd.DataFrame = rand_df.reindex(df.index)
rand_df
```

```python
# 포워드 필(forward fill)
rand_df_ffill : pd.DataFrame = rand_df.ffill()
rand_df_ffill
```

```python
# 이동평균 moving average, 롤링 평균 또는 중앙값
rand_df_rolling : pd.DataFrame = rand_df.rolling(window=len(rand_df), min_periods=1).mean()
rand_df_rolling
```
